```python
from bs4 import BeautifulSoup
import numpy as np
import pandas as pd 
import requests
import csv

source = requests.get ('https://en.wikipedia.org/wiki/List_of_postal_codes_of_Canada:_M').text
soup = BeautifulSoup(source,'lxml')
table = soup.find('table')

# extract table body
tablebody = table.tbody.text

# Save table body text to file
text_file = open("TableBody.txt", "w")
text_file.write(tablebody)
text_file.close()

# Read file to variable
text_file = open("TableBody.txt", "r")
TableBodyText=text_file.readlines()
text_file.close()

#parse through table body text and reformat to CSV file
count=0
record=""
field2=""
with open('tableBody.csv', 'w') as f:
    writer = csv.writer(f)
    for dataLine in TableBodyText: 
        if dataLine == '\n': 
            print("nothing" + " "+ dataLine)
                
        elif count == 0:
                    record = dataLine.rstrip('\n')
                    count=count+1
                    print("Count 1" + record)
                    
        elif count==1:
                    if dataLine.rstrip('\n')=='Not assigned':
                        print("Not assigned")
                        count = count+2
                    else:
                        field2 = dataLine.rstrip('\n')
                        record = record +','+ field2
                        count=count+1
                        print("Count 2B" + record)
                        
        elif count==2:
                if dataLine.rstrip('\n')=='Not assigned' and field2 == "Not assigned":
                    print("Count 3A" + record)
                    count = 3
                elif dataLine.rstrip('\n')=='Not assigned':
                    record = record +','+ field2
                    print("Count 3B" + record)
                    f.write(record +'\n')
                    print("Write Record")
                    count= 0
                    record = ''
                    field2 = ''                    
                    
                else: 
                    record = record +','+ dataLine.rstrip('\n')
                    print("Count 3C" + record)
                    f.write(record +'\n')
                    print("Write Record")
                    count= 0
                    record = ''
                    field2 = '' 
                
        elif count==3:
                count= 0
                record = ''
                field2 = '' 
                pass
 
    
#f.close

df_TorontoData = pd.read_csv("tableBody.csv") 
df_TorontoData
df_TorontoData.shape
```
